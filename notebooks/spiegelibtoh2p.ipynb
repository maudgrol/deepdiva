# SpiegeLib Array to h2p

```python
# parameter_dictionary = {0: {"parameter_index": 0, "value": "Diva", "label": "#AM"}}
header_dictionary = {
    "Bank": "Diva Factory Best Of", 
    "Author": "Bronto Scorpio", 
    "Description": "push the modulation wheel", 
    "Usage": "MW = filter\\r\\nAT = vibrato"
}
```

```python
import json
with open("h2p_dict.json") as json_file:
    parameter_dictionary = json.load(json_file)
parameter_dictionary
```

```python
def spiegelib_data_to_h2p(header_dictionary, parameter_dictionary, filename):
    
    with open(f"{filename}.h2p", "w") as f:

        # write header
        f.write("/*@Meta")
        f.write("\n\n")
        for header in header_dictionary.keys():
            f.write(f"{header}:")
            f.write("\n")
            f.write(f"'{header_dictionary[header]}'")
            f.write("\n\n")
        f.write("*/")
        f.write("\n\n")
    
        # write parameters
        for index in parameter_dictionary.keys():
            label = parameter_dictionary[index]["label"]
            value = parameter_dictionary[index]["value"]
            f.write(f"{label}={value}")
            f.write("\n")
```

```python
spiegelib_data_to_h2p(header_dictionary, parameter_dictionary, "filename")
```
